# 3. 분양가격 컬럼 살펴보기

```python
# NaN값이 있는 것도 확인. 데이터 타입도 문자열(object) 타입인 것을 확인
#  왜???
# for value in df['분양가격']:
#   print(value)
df['분양가격'].astype(int)
```

```python
# 공백 제거
df['분양가격'] = df['분양가격'].str.strip()
df.loc[df['분양가격'] == '']
```

```python
# 비어있는 분양가격을 0으로 넣기
df.loc[df['분양가격']=='', '분양가격'] = 0
```

```python
df['분양가격'].astype(int)
```

```python
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)
```

```python
df['분양가격'] = df['분양가격'].str.replace(",", "")
```

```python
df['분양가격'].astype(int)
```

```python
df['분양가격'] = df['분양가격'].fillna(0)
df['분양가격'].astype(int)
```

```python
df['분양가격'] = df['분양가격'].str.replace("-","")
df['분양가격'] = df['분양가격'].fillna(0)
```

```python
df['분양가격'].astype(int)
```

```python
df.loc[df['분양가격']=='', '분양가격'] = 0
```

```python
df['분양가격'] = df['분양가격'].astype(int)
df.info()
```

# "전용면적" 문자열 제거

```python
df['규모구분'] = df['규모구분'].str.replace("전용면적 ", "")
df['규모구분'].value_counts()
```

# 지역 별 평균 분양가격 확인하기

```python
df.groupby("지역명")["분양가격"].mean()
```

```python
# 분양가격이 0인 행은 제외하고 집계
mask = df['분양가격'] > 0
df.loc[mask].groupby("지역명")["분양가격"].mean()
```

특정 조건에 만족하는 행을 제거하고자 할 때
 1. index를 list로 가져온다.
 2. drop을 활용하여 행을 제거 한다.

```python
idx = df.loc[df["분양가격"] == 0].index
idx
```

```python
df_0_drop = df.drop(idx, axis=0)
```

```python
df_0_drop.groupby("지역명")["분양가격"].mean()
```

```python
(df_0_drop.groupby("지역명")["분양가격"].max() - df_0_drop.groupby("지역명")["분양가격"].min()).sort_values(ascending=False)
```

```python
df_0_drop.groupby("연도")["분양가격"].mean()
```

```python
# 연도별 규모구분 마다의 분양가격 평균 집계
df_0_drop.groupby(["연도", "규모구분"])["분양가격"].mean()
```

```python
pd.pivot_table(
    df_0_drop,
    index="연도",
    columns="규모구분",
    values="분양가격",
    # aggfunc="mean",
    margins=True
)
```
